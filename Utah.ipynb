```python
import pandas as pd
import hashlib
from fuzzywuzzy import fuzz
import Levenshtein

def calculate_hash(url):
    # Calculate hash of the URL
    url_hash = hashlib.md5(url.encode()).hexdigest()
    return url_hash

def find_duplicate_urls_hash(urls):
    duplicate_urls = []

    # Create a dictionary to store URL hashes
    url_hashes = {}

    for url in urls:
        # Calculate hash of the URL
        url_hash = calculate_hash(url)

        # Check for duplicate using hash
        if url_hash in url_hashes:
            duplicate_urls.append((url, url_hashes[url_hash]))
        else:
            url_hashes[url_hash] = url

    return duplicate_urls

def find_duplicate_urls_fuzzy(urls, threshold=93):
    duplicate_urls = []

    for i in range(len(urls)):
        for j in range(i + 1, len(urls)):
            # Calculate fuzzy ratio of the URLs
            fuzzy_ratio = fuzz.ratio(urls[i], urls[j])

            # Check for duplicate using fuzzy ratio
            if fuzzy_ratio >= threshold:
                duplicate_urls.append((urls[i], urls[j]))

    return duplicate_urls

def find_duplicate_urls_levenshtein(urls, threshold=2):
    duplicate_urls = []

    for i in range(len(urls)):
        for j in range(i + 1, len(urls)):
            # Calculate Levenshtein distance between the URLs
            distance = Levenshtein.distance(urls[i], urls[j])

            # Check for duplicate using Levenshtein distance
            if distance <= threshold:
                duplicate_urls.append((urls[i], urls[j]))

    return duplicate_urls

# Read data from Excel file
df = pd.read_excel('/content/UtahBommersHub.xlsx')

# Extract URLs from the dataframe
urls = df['Property'].tolist()

# Find duplicate URLs using hash algorithm
duplicate_urls_hash = find_duplicate_urls_hash(urls)

# Find duplicate URLs using fuzzy algorithm
duplicate_urls_fuzzy = find_duplicate_urls_fuzzy(urls)

# Find duplicate URLs using Levenshtein algorithm
duplicate_urls_levenshtein = find_duplicate_urls_levenshtein(urls)

# Print results
total_urls = len(urls)
total_duplicates_hash = len(duplicate_urls_hash)
total_duplicates_fuzzy = len(duplicate_urls_fuzzy)
total_duplicates_levenshtein = len(duplicate_urls_levenshtein)

print("Total URLs: ", total_urls)

print("Duplicate URLs (Hash Algorithm): ", total_duplicates_hash)
if duplicate_urls_hash:
    for url_pair in duplicate_urls_hash:
        print(url_pair[0])
        print(url_pair[1])
        print()

print("Duplicate URLs (Fuzzy Algorithm): ", total_duplicates_fuzzy)
if duplicate_urls_fuzzy:
    for url_pair in duplicate_urls_fuzzy:
        print(url_pair[0])
        print(url_pair[1])
        print()

print("Duplicate URLs (Levenshtein Algorithm): ", total_duplicates_levenshtein)
if duplicate_urls_levenshtein:
    for url_pair in duplicate_urls_levenshtein:
        print(url_pair[0])
        print(url_pair[1])
        print()
```
